## 1 sum the tho smallest numbers

Create a function that returns the sum of the two lowest positive numbers given an array of minimum 2 integers. No floats or empty arrays will be passed.

```python
# my solution
from typing import List

def sum_two_smallest_numbers(integers):
  sort=sorted(integers)    
  return sum (sort[0:2])
```

```python
# AI solution

from typing import List

def sum_two_smallest_numbers(ns: List[int]) -> int:
    return sum(list(sorted(ns))[:2])
```

## 2 find the parity outlier

You are given an array (which will have a length of at least 3, but could be very large) containing integers. The array is either entirely comprised of odd integers or entirely comprised of even integers except for a single integer N. Write a method that takes the array as an argument and returns this "outlier" 

```python
# my solutions
# solution 1
def find_outlier(integers):
  odd=[]
  even=[]
  for integer in integers: 
    if integer%2 != 0: 
      odd.append(integer)
    else: 
      even.append(integer)
  if len(odd)==1: 
    return odd[0]
  if len(even)==1:
    return even[0]
```

```python
# solution 2: 
def find_outlier(integers): 
    parity = sum([i%2 for i in integers[0:3]])
    for i in integers: 
        if parity > 1:
            if i%2 ==0:  
                return i
        else: 
            if i%2 ==1: 
                return i
            
# more compact
def find_outlier(integers): 
    parity = sum([i%2 for i in integers[0:3]])
    for i in integers: 
        if (parity>1 and i%2==0) or (parity <=1 and i%2 ==1):
            return i
```

```python
# solution 3: 
def find_outlier(integers):
    for i in range(-1,len(integers)-1):
        if (integers[i]%2 != integers[i-1]%2 and integers[i]%2 != integers[i+1]%2):
            return integers[i]
    
```

```python
# AI solution

from typing import List

def find_outlier(integers: List[int]) -> int:
    evens = [x for x in integers if x % 2 == 0]
    odds = [x for x in integers if x % 2 != 0]
    return odds[0] if len(evens) > 1 else evens[0]
```

## 3 Merge strings

Your function must merge strings a and b, and then return a single merged string. A merge operation on two strings is described as follows:

Append alternating characters from a and b, respectively, to some new string, mergedString.
Once all of the characters in one of the strings have been merged, append the remaining characters in the other string to mergedString.

```python
#Â my solution
def merge_strings(a, b):
  new_str = ""
  for i in range(min(len(a), len(b))):
    new_str += a[i]+b[i]
  if len(a)<len(b):
    new_str += b[len(a):]
  if len(b)<len(a):
    new_str += a[len(b):]
  return new_str
```

```python
# AI solution
def merge_strings(a: str, b: str) -> str:
    res = []
    i = 0
    j = 0

    while i < len(a) or j < len(b):
        if i < len(a):
            res.append(a[i])
            i += 1
        if j < len(b):
            res.append(b[j])
            j += 1

    return "".join(res)
```

## 4 Highest and Lowest
In this little assignment you are given a string of space separated numbers, and have to return the highest and lowest number.

```python
# my solution
def high_and_low(numbers):
  numeric=[int(i) for i in numbers.split()]
  return f"{max(numeric)} {min(numeric)}"
```

```python
# AI solution
def high_and_low(numbers: str) -> str:
    nn = [int(s) for s in numbers.split(" ")]
    return "%i %i" % (max(nn),min(nn))
```

## 5 First non repeating letter

Write a function that takes a string input,and returns the first character that is not repeated anywhere in the string. Characters in strings consist of printable characters.

As an added challenge, upper- and lowercase letters are considered the same character, but the function should return the correct case for the initial letter. For example, the input 'sTreSS' should return 'T'.

If a string contains all repeating characters, it should return the empty string ("").



```python

def first_non_repeating_letter(s):
  for letter in s: 
    if s.lower().count(letter.lower()) == 1:
      return letter
  return ""
```

```python
# AI solution
def first_non_repeating_letter(s: str) -> str:
    for c in s:
        if s.lower().count(c.lower()) == 1:
            return c
    return ""
```
